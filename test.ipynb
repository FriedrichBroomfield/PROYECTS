# **Boolean logic - Examples**



```python
a = False
b = True

print("a: ",a)
print("b: ",b)
print("")
print("mult. logic: ",bool(a*b))
print("sum. logic: ",bool(a+b))
print("")
print("and. logic: ",a and b)
print("or. logic: ",a or b)

# ----------------------------

# AND gate: logical AND
print("\nAND GATE")
for i in [False, True]:
    for j in [False, True]:
        print(f"{i} AND {j} = {i and j}")

# Or gate: inclusive OR
print("\nOR GATE")
for i in [False, True]:
    for j in [False, True]:
        print(f"{i} OR {j} = {i or j}")

# NOT gate: negation
print("\nNOT GATE")
for i in [False, True]:
    print(f"NOT {i} = {not i}")

# IMPLY gate: implication
print("\nIMPLY GATE")
for i in [False, True]:
    for j in [False, True]:
        print(f"{i} IMPLY {j} = {not i or j}")

# NIMPLY gate: inverse of IMPLY
print("\nNIMPLY GATE: Negation of IMPLY")

# XOR gate: exclusive OR
print("\nXOR GATE")
for i in [False, True]:
    for j in [False, True]:
        print(f"{i} XOR {j} = {i != j}")

# NAND gate: inverse of AND
print("\nNAND GATE: Negation of AND")

# NOR gate: inverse of OR
print("\nNOR GATE: Negation of OR")

# XNOR gate: inverse of XOR
print("\nXNOR GATE")
for i in [False, True]:
    for j in [False, True]:
        print(f"{i} XNOR {j} = {not (i != j)}")

# ---------------------------- LOGIC GATES using math operators ---------------------------

# Boolean values as integers
print("\nBoolean values as integers:")
for i in [0, 1]:
    print(f"{bool(i)} as integer: {i}")

# AND as MULTIPLICATION
print("\nAND as MULTIPLICATION")
for i in [0, 1]:
    for j in [0, 1]:
        print(f"{bool(i)} * {bool(j)} = {i * j} (interpreted as {bool(i * j)})")

# OR as ADDITION
print("\nOR as ADDITION")
for i in [0, 1]:
    for j in [0, 1]:
        print(f"{bool(i)} + {bool(j)} = {min(i + j, 1)} (interpreted as {bool(min(i + j, 1))})")

# NOT as SUBTRACTION from 1
print("\nNOT as SUBTRACTION from 1")
for i in [0, 1]:
    print(f"1 - {bool(i)} = {1 - i} (interpreted as {bool(1 - i)})")


```
